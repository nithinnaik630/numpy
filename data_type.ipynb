```python
# array search
```

```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5, 6, 4, 4])
new_arr = np.where(arr == 4)
print(new_arr)
```

```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5, 6, 4, 4])
new_arr = np.where(arr % 3)
print(new_arr)
```

```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5, 6, 4, 4])
new_arr = np.where(arr % 3==1)
print(new_arr)
```

```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5, 6, 4, 4])
new_arr = np.searchsorted(arr, 5)
print(new_arr)
```

```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5, 6, 4, 4])
new_arr = np.searchsorted(arr, 2,side='left')
print(new_arr)
```
