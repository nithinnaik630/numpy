```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# simulate 5000 trials
samples = np.random.binomial(n=20, p=0.7, size=5000)

# histogram
plt.hist(samples, bins=range(0, 22), edgecolor='black', alpha=0.7)

# labels and title
plt.xlabel("Number of students passed")
plt.ylabel("Frequency (out of 5000 trials)")
plt.title("Histogram of Exam Pass Counts (n=20, p=0.7)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Problem 1: Generate samples
samples = np.random.uniform(0, 10, 1000)

# Histogram
plt.hist(samples, bins=20, density=True, edgecolor='black')
plt.title("Uniform(0,10) Distribution")
plt.show()

# Mean and variance
print("Sample mean:", np.mean(samples))
print("Sample variance:", np.var(samples))
```

```python
# Rayleigh distribution
```

```python
from numpy import random

x = random.rayleigh(scale = 2, size = (2, 3))

print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.rayleigh(scale = 2, size = 1000),kind = "kde")

plt.title('rayleigh distribution')
plt.xlabel('magnituide')
plt.ylabel('density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 samples from a Rayleigh distribution with a scale of 2
rayleigh_samples = np.random.rayleigh(scale=2, size=1000)

# Visualize the distribution (optional)
plt.hist(rayleigh_samples, bins=50, density=False, alpha=0.7, color='orange')
plt.title('Rayleigh Distribution Samples')
plt.xlabel('Magnitude')
plt.ylabel('Probability Density')
plt.show()
```
