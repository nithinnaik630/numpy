```python
import numpy as np

data = np.genfromtxt('data.csv',delimiter=',', skip_header=1, filling_values = 1)

mean_val = np.mean(data)
medium_val = np.median(data)
std_val = np.std(data)

var_val = np.var(data)

min_val = np.min(data)
max_val = np.max(data)
range_val = np.ptp(data)


print(mean_val)
print(medium_val)
print(std_val)
print(var_val)

print(min_val)
print(max_val)
print(range_val)
```

```python
import cv
import numpy as np

# Initialize webcam capture
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur
    blurred = cv2.GaussianBlur(gray, (7, 7), 0)

    # Edge detection using Canny
    edges = cv2.Canny(blurred, 50, 150)

    # Create a color mask (e.g., filter out everything but red hues)
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    lower_red = np.array([0, 120, 70])
    upper_red = np.array([10, 255, 255])
    mask1 = cv2.inRange(hsv, lower_red, upper_red)
    lower_red2 = np.array([170, 120, 70])
    upper_red2 = np.array([180, 255, 255])
    mask2 = cv2.inRange(hsv, lower_red2, upper_red2)
    red_mask = mask1 + mask2

    # Bitwise-AND mask and original image
    red_filtered = cv2.bitwise_and(frame, frame, mask=red_mask)

    # Show the original frame and processed frames
    cv2.imshow('Original', frame)
    cv2.imshow('Edges', edges)
    cv2.imshow('Red Filtered', red_filtered)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
