```python
# LCM
```

```python
import numpy as np

num1 = 3
num2 = 4

x = np.lcm(num1, num2)

print(x)
```

```python
import numpy as np

x = np.array([2, 3, 4])

x1 = np.lcm.reduce(x)

print(x1)

```

```python
import numpy as np

x = np.array([1, 11])

x1 = np.lcm.reduce(x)

print(x1)
```

```python
# GCM
```

```python
import numpy as np

num1 = 2
num2 = 3

x = np.gcd(num1, num2)

print(x)
```

```python
import numpy as np

num = np.array([11, 40, 5])

x = np.gcd.reduce(num)

print(x)
```

```python
import numpy as np

num = np.array([11, 40, 5, 39, 28, 10])

x = np.gcd.reduce(num)

print(x)
```

```python
# FINDING UNION
```

```python
import numpy as np

num1 = np.array([1, 2, 3, 4])
num2 = np.array([5, 6, 7, 8])

x = np.union1d(num1, num2)

print(x)
```

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])

newarr = np.intersect1d(arr1, arr2, assume_unique=True)

print(newarr)
```
