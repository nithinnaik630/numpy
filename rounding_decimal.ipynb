```python
# Rounding()
```

```python
import numpy as np

x = np.around([12.5333, -2.334, -1.2])

print(x)
```

```python
import numpy as np

x = np.round([12.5333, -2.334, -1.2])

print(x)
```

```python
import numpy as np

x = np.round([12.5333, -2.534, -1.5])

print(x)
```

```python
import numpy as np

x = np.array([12.5333, -2.334, -1.2])

# 2 decimal number
new_value = np.around(x, 2)

# 3 decimal number
arr = np.around(x, 3)

# nearest value
arr1 = np.around(x)

print(new_value)
print(arr)
print(arr1)
```

```python
# floor()
```

```python
import numpy as np

x = np.floor([12.5333, -2.334, -1.2])

print(x)
```

```python
import numpy as np

x = np.array([12.5333, -2.334, -1.2])

x1 = np.around(np.add(x,2.5))

arr = np.floor(x)

print(arr)
print(x1)
```

```python
# Ceil()
```

```python
import numpy as np

x = np.ceil([1.3333, 2.0, 1.5555])

print(x)
```

```python
import numpy as np

x = np.ceil([1.3333, 2.0, -1.5555])

print(x)
```

```python
import numpy as np

arr = np.array([2.333, 2.0, 3.5, -2.8])

# truncation
arr1 = np.trunc(arr)

# fix
arr2 = np.fix(arr)

# round
arr3 = np.round(arr)

# floor
arr4 = np.floor(arr)

#ceil
arr5 = np.ceil(arr)

print(arr1)
print(arr2)
print(arr3)
print(arr4)
print(arr5)
```

```python
# Summations
```

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 9])

print(np.sum([arr1, arr2]))
```

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4])
# arr2 = np.array([5, 6, 7, 9])

print(np.cumsum([arr1]))
```

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 9])

print(np.cumsum([arr1, arr2]))
```

```python
# product 
```

```python
import numpy as np

vec1 = np.array([1, 2, 5, 9])
vec2 = np.array([2, 8, 5, 3])  # clacuclting the multplition of vec1 and vec2

new_vec = np.prod([vec1, vec2])

print("multipling vec_total:", new_vec)
```

```python
import numpy as np

vec1 = np.array([[1, 2],[ 5, 9]])
vec2 = np.array([[2, 8], [5, 3]])  # clacuclting the multplition of vec1 and vec2 using 2 dimension array

new_vec = np.prod([vec1, vec2])

print(new_vec)
```

```python
import numpy as np

vec1 = np.array([1, 2, 5, 9])
vec2 = np.array([2, 8, 5, 3])  # clacuclting the multplition of vec1 and vec2

new_vec = np.prod([vec1, vec2], axis=0) # multipling coloums

print(new_vec)
```
